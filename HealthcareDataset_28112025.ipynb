# âœ… What does Skewness Mean?

Skewness tells you:

"Is the data symmetric or tilted to one side?"

âœ” Positive skew (+)

â†’ Long tail on the right
â†’ More small values, few large values

âœ” Negative skew (â€“)

â†’ Long tail on the left
â†’ More large values, few small values

âœ” Near 0

â†’ Almost perfectly symmetric (balanced)

â­ Your Skewness Value = -0.0065

This is VERY close to 0
(0.00 would be perfect symmetry)

```python
df_healthcare['LOS']=(df_healthcare['Discharge Date']-df_healthcare['Date of Admission']).dt.days
df_healthcare['LOS'].describe()
```

```python
#create stay categories
bins = [0, 3, 7, float('inf')]
labels = ['Short Stay', 'Medium Stay', 'Long Stay']

df_healthcare.loc[:,'Stay_Category'] = pd.cut(df_healthcare['LOS'], bins=bins, labels=labels, right=True)
```

```python
bins = [0, 18, 30, 45, 60, 75, 100]
labels=['0-18', '19-30', '31-45', '46-60', '61-75', '76+']
df_healthcare['Age_Group'] = pd.cut(df_healthcare['Age'], bins=bins, labels=labels, right=True)
df_healthcare['Age_Group'].value_counts()
```



1.   It creates these intervals:

0 â†’ 18

18 â†’ 30

30 â†’ 45

45 â†’ 60

60 â†’ 75

75 â†’ 100

Think of bins as cut points
2.  Ages 0â€“18 â†’ labeled "0-18"

Ages 19â€“30 â†’ labeled "19-30"

Ages 31â€“45 â†’ labeled "31-45"

etc.

3. pd.cut() does:

âœ” Splits the Age column into groups
âœ” Assigns labels based on which bin the age belongs to
4.   parameter right=True means:

âž¡ The right edge (upper limit) is included
Example:
18 goes into 0â€“18,

# Billing per day (if LOS > 0)

```python
df_healthcare.loc[:, 'LOS'] = df_healthcare['LOS'].replace(0, 1)
df_healthcare.loc[:, 'Bill_per_Day'] = df_healthcare['Billing Amount'] / df_healthcare['LOS']
```

```python
df_healthcare['Billing Category']=pd.qcut(df_healthcare['Billing Amount'],3,labels=['Low','Medium','High'])
```

```python
df_healthcare.loc[:, 'Date of Admission'] = pd.to_datetime(
    df_healthcare['Date of Admission'], errors='coerce'
)
```

```python
df_healthcare.loc[:, 'Admission_Week'] = df_healthcare['Date of Admission'].dt.isocalendar().week
```

```python
df_healthcare.loc[:,'Admission Day']=df_healthcare['Date of Admission'].dt.day_name()
```

```python
df_healthcare.loc[:, 'Admission_Month'] = df_healthcare['Date of Admission'].dt.month
```

```python
season_map = {
    12:'Winter', 1:'Winter', 2:'Winter',
    3:'Spring', 4:'Spring', 5:'Spring',
    6:'Summer', 7:'Summer', 8:'Summer',
    9:'Autumn', 10:'Autumn', 11:'Autumn'
}

df_healthcare.loc[:, 'Season_of_Admission'] = df_healthcare['Admission_Month'].map(season_map)
```

```python
#Weekend Admission (Yes/No)
df_healthcare.loc[:,'Weekend_Admission'] = df_healthcare['Admission Day'].isin(['Saturday','Sunday']).astype(int)
```

```python
df_healthcare.head(4)
```

```python
df_healthcare['Admission_Quarter'] = df_healthcare['Date of Admission'].dt.quarter
```

```python
#Total Days in Hospital per Year (grouped patient-wise if ID existed)
```

```python
df_healthcare['Original_Billing'] = df_healthcare['Billing Amount'] * 1.20
```

```python
df_healthcare['Discount_Percentage'] = (
    (df_healthcare['Original_Billing'] - df_healthcare['Billing Amount'])
    / df_healthcare['Original_Billing'].replace(0, 1)
) * 100
```

```python
df_healthcare['Senior_Citizen'] = (df_healthcare['Age'] >= 60).astype(int)
```

```python
df_healthcare['Chronic_Disease'] = df_healthcare['Medical Condition'].isin(['Diabetes','Cancer','Hypertension']).astype(int)
```

```python
df_healthcare['Is_Emergency'] = (df_healthcare['Admission Type']=='Emergency').astype(int)
```

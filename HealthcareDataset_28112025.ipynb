# ✅ What does Skewness Mean?

Skewness tells you:

"Is the data symmetric or tilted to one side?"

✔ Positive skew (+)

→ Long tail on the right
→ More small values, few large values

✔ Negative skew (–)

→ Long tail on the left
→ More large values, few small values

✔ Near 0

→ Almost perfectly symmetric (balanced)

⭐ Your Skewness Value = -0.0065

This is VERY close to 0
(0.00 would be perfect symmetry)

# Hypothesis testing helps us decide whether a pattern in data is REAL or just RANDOM.
# ⭐ 1. To Compare Groups (Means or Proportions)

Example questions:

Do male and female patients have different billing?

Do patients in Hospital A stay longer than Hospital B?

Does Diabetes cost more than Cancer?

Without hypothesis testing → we’re guessing
With hypothesis testing → we KNOW

# ⭐ 2. To Check If a Strategy or Change Had an Effect

Examples:

New medicine reduces stay time or not?

New pricing plan increases revenue or not?

New website design increases conversion rate or not?

# ⭐ 3. To Make Data-Driven Decisions

Companies use hypothesis testing to avoid wrong decisions.

Example:

A manager sees:

Male billing: ₹25,500
Female billing: ₹25,480


Difference = only ₹20
Is this REAL or RANDOM?

Hypothesis testing tells:

Difference is NOT statistically significant
→ So treat both groups the same
→ No need for separate pricing strategy

# 🔵 3️⃣ ttest_ind(groupA, groupB, equal_var=False)

This function calculates:

t-statistic → how different the means are

p-value → is the difference statistically meaningful or just random?

equal_var=False means:

→ We assume the variances of two groups may not be equal
(This is called Welch's t-test, safer and recommended)

# ✅ 1. First, define H₀ and H₁

For gender vs billing:

H₀ (Null Hypothesis)

👉 There is NO difference in average billing between Male and Female.

H₁ (Alternative Hypothesis)

👉 There IS a difference in average billing between Male and Female.

You ALWAYS compare p-value with α (0.05).

# ✅ 3. Decision Rule

Use:

if p < 0.05  → Reject H₀ (significant difference)
if p > 0.05  → Fail to reject H₀ (NO significant difference)

# ✅ 1. Billing Amount — Primary Fraud Indicator

Fraudsters inflate billing.

🔍 Fraud Clues from Billing Amount:

Very high billing (top 1% or Z-score > 3)

Sudden spike compared to same condition

High billing for normal / simple tests

Large bill but short stay (short-stay high billing)

📌 Why important?
Most fraud patterns start from abnormally high bills.


✅ 2. Medical Condition — Cost should match severity

Each condition has a predictable cost pattern.

🔍 Fraud Clues:

High billing for simple conditions
(ex: "Allergy" → ₹40,000 is suspicious)

Low billing for serious conditions
(means undercoding / manipulation)

Rare diseases used repeatedly → fraudulent claims

📌 Why?
Fraud happens when billing does not match condition severity.


✅ 3. Admission Type — Mismatch Detection

Admission Types:

Emergency → usually high cost

Elective (Planned) → moderate

Urgent → medium

🔍 Fraud Clues:

Elective case has Emergency-level billing

Urgent admission but very low bill

Emergency admission with no abnormal tests

📌 Why?
Fraudsters often mark planned procedures as "Emergency" to charge more.


✅ 4. Doctor — Provider Fraud / Overbilling

Each doctor has a pattern of billing.

🔍 Fraud Clues:

Doctor consistently producing high bills

Doctor seeing more high-cost patients than others

Doctor with too many short-stay high-billing cases

Doctor involved in repeated high-value normal-test claims

📌 Why?
Provider-level fraud is common in healthcare.


✅ 5. Hospital — Institution-level Fraud

Some hospitals show abnormal behavior.

🔍 Fraud Clues:

Hospitals with highest average bill

Hospitals with extremely high bill per day

Hospitals with abnormal number of short stays

Hospitals generating more “Emergency” cases than average

📌 Why?
Some hospitals inflate bills.

✅ 6. Test Results — Mismatch with Billing

This is one of the strongest indicators of fraud.

🔍 Fraud Clues:

Normal or Inconclusive test results but VERY high billing

Abnormal test results but low billing (under-reporting)

No tests but treatment billed

📌 Why?
Billing should correlate with diagnosis & test results.
